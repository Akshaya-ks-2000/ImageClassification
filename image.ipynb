```python
import pandas as pd
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage.io import imread
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
import numpy as np
```

resize: to make all the images of same size

imread: image reading

```python
data_array=[]
target_array=[]
Categories=['Cat','Dog']
datadir='/content/drive/MyDrive/DATA'
```

```python

for i in Categories:
  sub_path=os.path.join(datadir,i)
  images=os.listdir(sub_path)
  # print(images)
  for img in images:
    img_path=os.path.join(sub_path,img)
    img_array=imread(img_path)
    img_resized=resize(img_array,(150,150,3))
    data_array.append(img_resized.flatten())
    target_array.append(Categories.index(i))
  print('loaded category :',i)
```

```python
data=np.array(data_array)
data.shape           #150*150*3=67500
```

```python
target=np.array(target_array)
target.shape
```

```python
df=pd.DataFrame(data)
df
```

here values are normalised during resizeing

```python
df['target']=target
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
sv=SVC(kernel='poly')
sv.fit(X_train,y_train)
y_pred=sv.predict(X_test)
y_pred
print(classification_report(y_pred,y_test))
print(ConfusionMatrixDisplay.from_predictions(y_pred,y_test))
```
